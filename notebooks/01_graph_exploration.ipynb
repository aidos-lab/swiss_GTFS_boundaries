# 01 — Graph Exploration

Load a pre-built filtered GTFS zip, construct the transit graph, and render an interactive map.

**Prerequisites:** Run `gtfs-filter` and optionally `gtfs-build` first, or ensure filtered GTFS zips exist under `Data/{scale}/{city}.zip`.

```python
import sys
sys.path.insert(0, '../src')

import gtfs2nx as gx

from swiss_gtfs.graphs.build import build_summary_graph
from swiss_gtfs.graphs.visualize import visualize_graph
```

```python
# --- Configuration ---
SCALE = 'agglomeration'
CITY  = 'bern'
GTFS_DIR = '../Data'

gtfs_zip = f'{GTFS_DIR}/{SCALE}/{CITY}.zip'
```

```python
# Build the graph using city2graph
nodes_gdf, edges_gdf = build_summary_graph(
    gtfs_zip,
    start_time='07:00:00',
    end_time='10:00:00',
)
print(f'Graph: {len(nodes_gdf)} nodes, {len(edges_gdf)} edges')
```

```python
# Build a gtfs2nx graph for legacy visualisation (uses stop-level transit graph)
G_nx = gx.transit_graph(gtfs_zip, time_window=('07:00', '10:00'))
print(f'gtfs2nx graph: {G_nx.number_of_nodes()} nodes, {G_nx.number_of_edges()} edges')
```

```python
# Render interactive folium map
m = visualize_graph(G_nx, gtfs_zip)
m
```
